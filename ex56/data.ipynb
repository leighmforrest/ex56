```python
import pandas as pd

pd.set_option('future.no_silent_downcasting', True)

# Load the Excel file
ttb = pd.read_excel("data/Statistical_Report_Beer_October_2021.xlsx")

# Extract the rows with 'Reporting Period' and 'Report Date'
dates = ttb.iloc[[3,4], 0]

# Extract relevant data rows from the main dataset
df = ttb.iloc[7:19]
df.reset_index(drop=True, inplace=True)

# Select the columns you want to work with (excluding unwanted ones)
df = df.iloc[:, [1,2,3,4,5,6,7]]

# Clean up the header and fill missing values
new_header = df.iloc[0].fillna("")
df = df[1:].fillna(0)
df.columns = new_header
df.reset_index(drop=True, inplace=True)

# Handle a specific issue with row 9 (adjusting a column value)
df.iloc[9, 2] = df.iloc[9, 1]

# Clean up column names (removing newline characters and trimming)
df.columns = df.columns.str.replace("\n", " ", regex=True).str.strip()

# Drop unwanted columns
df = df.iloc[:, 2:]


# REPORTING DATES AND REPORTING PERIOD

# Extract and clean the 'Reporting Period' and 'Report Date' values
reporting_period_str = dates[dates.str.contains('Reporting Period')].str.split(":", expand=True)[1].str.strip()
report_date_str = dates[dates.str.contains('Report Date')].str.split(":", expand=True)[1].str.strip()

# Convert 'Reporting Period' to datetime, assuming the 1st of the month
reporting_period = pd.to_datetime(reporting_period_str + ' 1', format='%B %Y %d')

# Convert 'Report Date' to datetime
report_date = pd.to_datetime(report_date_str, format='%d%b%Y')

# Create a DataFrame for the clean date columns
clean_dates = pd.DataFrame({
    "Reporting Period": reporting_period.values,
    "Report Date": report_date.values
})

# Ensure that clean_dates is repeated for all rows in df if it contains only one row
if len(clean_dates) == 1:
    clean_dates = pd.concat([clean_dates]*len(df), ignore_index=True)


df.insert(1, 'Reporting Period', clean_dates['Reporting Period'])
df.insert(2, 'Report Date', clean_dates['Report Date'])

# Clean the value
# NOTE: iloc[X,Y] => X is the row, Y is the column
# => df.iloc[9,0] is already a string, so call the replace() method directly.
df.iloc[9,0] = df.iloc[9,0].replace("\n", " ")
df.iloc[[0,10], :]
```
